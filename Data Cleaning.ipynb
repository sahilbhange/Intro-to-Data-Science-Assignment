```python
# Removing the records which has no data for 'title_year','language','genres' and 'country'  recorded.
data = data.dropna(subset=['title_year'])
data = data.dropna(subset=['language'])
data = data.dropna(subset=['genres'])
data = data.dropna(subset=['country'])
 
data = data.dropna(subset=['actor_2_facebook_likes'])
data = data.dropna(subset=['facenumber_in_poster'])
data = data.dropna(subset=['actor_3_facebook_likes'])
data = data.dropna(subset=['director_facebook_likes'])

data = data.dropna(subset=['actor_1_facebook_likes'])
data = data.dropna(subset=['num_voted_users'])
data = data.dropna(subset=['num_user_for_reviews'])
data = data.dropna(subset=['num_critic_for_reviews'])
    
data.shape
```

```python
# Since number of reviews can not be float, fixixng its format and similarly for other fields
data.actor_2_facebook_likes = data.actor_2_facebook_likes.astype(np.int64)
data.facenumber_in_poster = data.facenumber_in_poster.astype(np.int64)
data.actor_3_facebook_likes = data.actor_3_facebook_likes.astype(np.int64)
data.director_facebook_likes = data.director_facebook_likes.astype(np.int64)
data.actor_1_facebook_likes = data.actor_1_facebook_likes.astype(np.int16)
data.num_voted_users = data.num_voted_users.astype(np.int64)
data.num_user_for_reviews = data.num_user_for_reviews.astype(np.int16)
data.title_year = data.title_year.astype(np.int8)
data.num_critic_for_reviews = data.num_critic_for_reviews.astype(np.int16)



#data.dtypes.num_critic_for_reviews
#data.dtypes
```
